```python
sayi1=float(input("Birinci sayıyı giriniz: "))
sayi2=float(input("İkinci sayıyı giriniz: "))

toplama=sayi1 + sayi2
cikarma=sayi1 - sayi2
carpma=sayi1 * sayi2
bolme=sayi1 / sayi2
tam_bolme=sayi1 // sayi2
mod=sayi1 % sayi2
us=sayi1 ** sayi2

print("Toplama:",toplama)
print("Çıkarma:",cikarma)
print("Çarpma:",carpma)
print("Bölme:",bolme)
print("Tam Bölme:",tam_bolme)
print("Mod Alma:",mod)
print("Üs Alma:",us)
```
